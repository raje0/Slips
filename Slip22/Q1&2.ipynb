```python

MAX = 1
MIN = -1

def alpha_beta_pruning(node, depth, alpha, beta, maximizing_player):
   
    if depth == 0 or is_terminal(node):
        return evaluate(node)

    if maximizing_player:
        max_eval = -float('inf')
        for child in generate_children(node, MAX):
            eval = alpha_beta_pruning(child, depth - 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break  
        return max_eval
    else:
        min_eval = float('inf')
        for child in generate_children(node, MIN):
            eval = alpha_beta_pruning(child, depth - 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break  
        return min_eval


def generate_children(node, player):
    return [node - 1, node + 1]  

def is_terminal(node):
    return node == 0  
def evaluate(node):
    return node 

if __name__ == "__main__":
    initial_node = 10 
    depth = 4  
    alpha = -float('inf')
    beta = float('inf')

    
    result = alpha_beta_pruning(initial_node, depth, alpha, beta, True)
    print(f"Best value found by Alpha-Beta Pruning: {result}")
```

```python
import random

class SimpleChatbot:
    def __init__(self):
        self.responses = {
            'hello': ["Hi!", "Hello!", "Hey there!", "Greetings!"],
            'how are you': ["I'm doing great, thank you!", "I'm fine, how about you?", "Doing well! How can I assist you today?"],
            'bye': ["Goodbye!", "See you later!", "Take care!"],
            'name': ["I am a chatbot!", "You can call me Chatbot!", "I am here to help you!"],
            'default': ["I'm sorry, I didn't understand that.", "Can you please rephrase?", "I don't quite get that."]
        }

    def get_response(self, user_input):

        user_input = user_input.lower()


        for key in self.responses:
            if key in user_input:
                return random.choice(self.responses[key])


        return random.choice(self.responses['default'])

def chat():
    chatbot = SimpleChatbot()
    print("Chatbot: Hello! How can I help you today?")
    
    while True:
        user_input = input("You: ")
        
        if 'bye' in user_input.lower():
            print("Chatbot:", chatbot.get_response(user_input))
            break
        

        print("Chatbot:", chatbot.get_response(user_input))

if __name__ == "__main__":
    chat()
```
