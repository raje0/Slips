```python
import random

def function(x):
    """The mathematical function to maximize: f(x) = -x^2 + 4x + 3"""
    return -x**2 + 4*x + 3

def hill_climbing(start, max_iterations, step_size=0.1):
    """Hill Climbing Algorithm to maximize the function."""
    
    current_x = start
    current_value = function(current_x)
    
    print(f"Starting at x = {current_x}, f(x) = {current_value}")
    
    for _ in range(max_iterations):
    
        neighbors = [current_x + step_size, current_x - step_size]
        
        neighbor_values = [function(neighbor) for neighbor in neighbors]
        
        max_value = max(neighbor_values)
        
        if max_value > current_value:
            best_neighbor_index = neighbor_values.index(max_value)
            current_x = neighbors[best_neighbor_index]
            current_value = max_value
            print(f"Moved to x = {current_x}, f(x) = {current_value}")
        else:

            print(f"Local maximum reached at x = {current_x}, f(x) = {current_value}")
            break
    
    return current_x, current_value

if __name__ == "__main__":
   
    start_point = random.uniform(-10, 10)  

    best_x, best_value = hill_climbing(start_point, max_iterations)

    print(f"\nBest solution found: x = {best_x}, f(x) = {best_value}")
```
