```python
class CollegeBot:
    def __init__(self, name, department, year, courses):
        self.name = name
        self.department = department
        self.year = year
        self.courses = courses

    def get_info(self, query):
        query = query.lower()
        
        if 'name' in query:
            return f"My name is {self.name}."
        elif 'department' in query:
            return f"I belong to the {self.department} department."
        elif 'year' in query:
            return f"I am in my {self.year} year of college."
        elif 'courses' in query:
            return f"I am currently taking the following courses: {', '.join(self.courses)}."
        else:
            return "Sorry, I didn't understand your query."

student = CollegeBot("John Doe", "Computer Science", "3rd", ["Data Structures", "Operating Systems", "Algorithms"])
print(student.get_info("What is your name?"))
print(student.get_info("What department are you in?"))
print(student.get_info("What courses are you taking?"))
```

```python
from collections import deque


def find_empty_space(puzzle):
    return puzzle.index(0)


def get_possible_moves(puzzle):
    possible_moves = []
    empty_pos = find_empty_space(puzzle)
    row, col = empty_pos // 3, empty_pos % 3
    
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  
    for dr, dc in directions:
        new_row, new_col = row + dr, col + dc
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_pos = new_row * 3 + new_col
            new_puzzle = puzzle[:]
           
            new_puzzle[empty_pos], new_puzzle[new_pos] = new_puzzle[new_pos], new_puzzle[empty_pos]
            possible_moves.append(new_puzzle)
    return possible_moves


def bfs(start, goal):
    queue = deque([(start, [])])  
    visited = set()
    visited.add(tuple(start))  
    
    while queue:
        current_state, path = queue.popleft()
       
        if current_state == goal:
            return path + [current_state]
  
        for move in get_possible_moves(current_state):
            if tuple(move) not in visited:
                visited.add(tuple(move))
                queue.append((move, path + [current_state]))
    
    return None 

start_state = [1, 2, 3, 4, 5, 6, 7, 8, 0]
goal_state = [1, 2, 3, 4, 5, 6, 7, 0, 8]

solution = bfs(start_state, goal_state)

if solution:
    print("Solution found:")
    for step in solution:
        print(step[:3])
        print(step[3:6])
        print(step[6:])
        print() 
else:
    print("No solution found.")
```
